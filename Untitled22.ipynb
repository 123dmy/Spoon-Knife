```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Load the Excel file with the URLs
df = pd.read_excel('/content/Input.xlsx')

# Loop through each row in the Excel file
for index, row in df.iterrows():
    url_id = row['URL_ID']
    url = row['URL']
    
    # Get the HTML content from the URL
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Extract the article title and text from the HTML using Beautiful Soup
    title = soup.find('title').text.strip()
    article = soup.find('article')
    if article is not None:
        article_text = '\n'.join([p.text.strip() for p in article.find_all('p')])
    else:
        # Handle case where <article> tag is not found
        article_text = ''
        for p in soup.find_all('p'):
            if p.parent.name != 'footer' and p.parent.name != 'header':
                article_text += p.text.strip() + '\n'
    
    # Save the article text in a text file with the URL_ID as its file name
    with open(f'{url_id}.txt', 'w', encoding='utf-8') as f:
        f.write(f'{title}\n\n{article_text}')
```
